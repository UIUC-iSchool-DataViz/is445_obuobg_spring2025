### Plotting just the "bad" parts of Stitch (but really he's more complicated than that!)

````python
#reds_bad_mask = im_data[:,:,0] == 255 # oops!  this is for the good parts of stitch!
reds_bad_mask = im_data[:,:,0] == 126 # oops!  this is for the good parts of stitch!
# [all rows (y), all columns (x), first color channel from RGBA]
````

````python
#reds_bad_mask
````

````python
im_data_bad = im_data.copy() # just to ensure we have a new image
im_data_bad[~reds_bad_mask] = 0
# "~" NOT --> "~reds_bad_mask" is saying NOT in the bad mask
# = 0 set everything to 0, including the opacity --> not-bad parts of the image see-through
````

````python
fig, ax = plt.subplots(figsize=(3,3))
ax.set_facecolor('gray')
ax.imshow(im_data_bad)
plt.show()
````

````python
np.unique(im_data.reshape(-1, im_data.shape[2]),axis=0)
````

````python
reds_good_mask = im_data[:,:,0] == 255
greens_good_mask = im_data[:,:,1] == 255
blues_good_mask = im_data[:,:,2] == 255
alphas_good_mask = im_data[:,:,3] == 255

# practice with boolean and -- ALL of the conditions need to be true
pixel_mask_good = reds_good_mask & greens_good_mask & blues_good_mask & alphas_good_mask
````

````python
im_data_good = im_data.copy() # just to ensure we have a new image
im_data_good[~pixel_mask_good] = 0
````

````python
fig, ax = plt.subplots(figsize=(3,3))
ax.set_facecolor('gray')
ax.imshow(im_data_good)
plt.show()
````

````python
n_good = len(im_data[pixel_mask_good])
n_good
````

````python
n_bad = len(im_data[reds_bad_mask])
n_bad
````

````python
n_bad/(n_bad+n_good)
````

````python
fig2, ax2 = plt.subplots(4,1,figsize=(10,8)) # can set 8 --> 15 for longer figure

ax2[0].hist(im2_data[:,:,0].flatten(), bins=256, color='red') # 0=red channel
ax2[1].hist(im2_data[:,:,1].flatten(), bins=256, color='green') # 1=green channel
ax2[2].hist(im2_data[:,:,2].flatten(), bins=256, color='blue') # 2=blue channel
#ax2[3].hist(im2_data[:,:,3].flatten(), bins=256, color='gray') # 3=alpha/opacity channel

# add some x/y labels
ax2[0].set_xlabel('Red channel bins')
ax2[1].set_xlabel('Green channel bins')
ax2[2].set_xlabel('Blue channel bins')
#ax2[3].set_xlabel('Alpha channel bins')
for a in ax2[:-1]: # don't do this for the last axes!
    a.set_ylabel('# of pixels in each bin')
    a.set_ylim(0, 20000)

ax2[3].imshow(im2_data)

fig2.tight_layout() # try to ensure no overlapping plots & labels
plt.show()
````
