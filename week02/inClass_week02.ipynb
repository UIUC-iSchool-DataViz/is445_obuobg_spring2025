```python
# how much square footage was acquired per year?
buildings.groupby('Year Acquired')['Square Footage'].sum()
```

```python
# plot the total square footage per year acquired
fig, ax = plt.subplots(figsize=(10,2))
gbs = buildings.groupby('Year Acquired')['Square Footage'].sum()
gbs.plot(ax=ax)
ax.set_yscale('log')
```
