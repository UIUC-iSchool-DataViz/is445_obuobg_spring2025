```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.')
ax.plot(stats['min'], marker='.')
ax.set_xlabel('Year (Acquired)')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', label='Max')
ax.plot(stats['min'], marker='.', label='Min')
ax.set_xlabel('Year (Acquired)')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', label='Max')
ax.plot(stats['min'], marker='.', label='Min')
ax.set_xlabel('Year (Acquired)')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
#plt.style.use('ggplot')
@ipywidgets.interact(style=plt.style.available)
def make_plot(style):
    with plt.style.context(style):
        # CMD + ]
        fig, ax = plt.subplots(figsize=(10,2))
        ax.plot(stats['max'], marker='.', label='Max')
        ax.plot(stats['min'], marker='.', label='Min')
        ax.set_xlabel('Year (Acquired)')
        ax.set_ylabel('Square Footage')
        ax.set_yscale('log')
        ax.legend()
        plt.show()
```
