````python
bins = np.linspace(df['Student_teacher_ratio'].min(), df['Student_teacher_ratio'].max(), 10)
````

````python
bins
````

````python
table = df.pivot_table(index='State', columns=pd.cut(df['Student_teacher_ratio'],bins), 
                       aggfunc='size')
````

````python
table
````

````python
import matplotlib.pyplot as plt
````

````python
fig, ax = plt.subplots(figsize=(10,8))
extent = [bins.min(),bins.max(), 0, len(table.index)] # xmin, xmax, ymin, ymax
ax.imshow(table.values, cmap='hot', interpolation='nearest', extent=extent)
ax.set_yticks(range(len(table.index)))
ax.set_yticklabels(table.index)
plt.show()
````
