```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
```

```python
buildings = pd.read_csv("https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/building_inventory.csv",
           na_values = {'Square Footage': 0,
                       'Year Acquired': 0,
                       'Year Constructed': 0}) 
```

```python
buildings.head()
```

```python
stats = buildings.groupby('Year Acquired')['Square Footage'].describe()
stats
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0)
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0)
ax.plot(stats['min'], marker='.', linewidth=1.0)
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.', linewidth=1.0, label='Max')
ax.plot(stats['min'], marker='.', linewidth=1.0, label='Min')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Square Footage')
ax.set_yscale('log')
ax.legend()
plt.show()
```

```python
# Step 3 -- add a decorator to our function
@ipywidgets.interact(colormap = plt.colormaps())
def make_mich_plot(colormap):
    plt.imshow(michigan, cmap=colormap)
    plt.colorbar()
    plt.show()
```

```python
my_colormap_list = ['terrain', 'magma', 'viridis']

@ipywidgets.interact(colormap = my_colormap_list)
def make_mich_plot(colormap):
    plt.imshow(michigan, cmap=colormap)
    plt.clim(-352, 352)
    plt.colorbar()
    plt.show()
```

```python
# Step 1 -- take log: hardcode a value

colormap='terrain'
take_log = True

if take_log:
    norm = matplotlib.colors.SymLogNorm(10)
else:
    norm = matplotlib.colors.Normalize() # default color normaliztion
    
plt.imshow(michigan, cmap=colormap, norm=norm)
plt.clim(-352, 352)
plt.colorbar()
plt.show()
```

```python
# Step 2 -- take log & choose colormap: make this a function

# colormap='terrain'
# take_log = True

def make_mich_plot(colormap, take_log):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10)
    else:
        norm = matplotlib.colors.Normalize() # default color normaliztion
        
    plt.imshow(michigan, cmap=colormap, norm=norm)
    plt.clim(-352, 352)
    plt.colorbar()
    plt.show()
```

```python
# Step 2.5

make_mich_plot('Reds', False)
```

```python
# Step 3 -- take log & choose colormap: decorate the function
@ipywidgets.interact(colormap = my_colormap_list, take_log=True)
def make_mich_plot(colormap, take_log):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10)
    else:
        norm = matplotlib.colors.Normalize() # default color normaliztion
        
    plt.imshow(michigan, cmap=colormap, norm=norm)
    plt.clim(-352, 352)
    plt.colorbar()
    plt.show()
```
